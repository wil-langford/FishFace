```python
#########################
#  INITIAL ENVIRONMENT  #
#########################

%run test.py

import ipythonsupport
reload(ipythonsupport)
import sys, os, time
import cv2
import IPython.display as ipd

ips = ipythonsupport.FFiPySupport()
ips.msg("Starting initial environment setup.")
sys.path.append(os.path.join('..','fishface'))
try:
    import capture, imageframe, hopper, poser
    ips.msg("Imported FishFace Libraries.")
except:
    ips.msg("Couldn't import the FishFace libraries.  You may need to reinstall FishFace.", ips.ERR)
ips.ephemeraDirectory()
ips.msg("Initial environment setup complete.", ips.FINAL)
```

```python
############
# SETTINGS #
############

# Can be 'IR' or 'visible'.
LIGHT_TYPE = 'IR'

# Can be any combination of letters, numbers and spaces.  No punctuation or other
# special characters.
EXPERIMENT_NAME = 'My Experiment'

# If this is None, then the experiment will start as soon as you execute the Data Collection cell below.
# You can also specify a specific start time in H:M:S format.  (e.g. "15:16:17" would be 17 seconds after 3:16pm)
START_TIME = None

# This is the number of seconds between data image captures.  It will be rounded
# to the nearest positive integer with a minimum value of 1.
INTERVAL = 5

# These settings control the end of the experiment.  If NUMBER_OF_DATA is None, then it is later calculated from
# INTERVAL and STOP_TIME. STOP_TIME is in the same H:M:S format as START_TIME above.  If NUMBER_OF_DATA is specified
# (as a positive integer,) then STOP_TIME is ignored.
NUMBER_OF_DATA = 10
STOP_TIME = None

######### The following can probably be left alone. #########

#DATA_DIR = os.path.join('.','data')
DATA_DIR = os.path.join('.','ephemera')
EXPERIMENT_START_TIME = ips.dtg()[:-2]
EXPERIMENT_DIR = os.path.join(DATA_DIR,
                                    "{}-{}".format(EXPERIMENT_NAME,
                                                   EXPERIMENT_START_TIME))
os.mkdir(EXPERIMENT_DIR)

CAL_IMAGE_PREFIX = 'CAL'
DATA_IMAGE_PREFIX = 'DATA'
DATA_SERIES = 0
```

```python
#############################
# CAPTURE CALIBRATION IMAGE #
#############################

calFilename = os.path.join(EXPERIMENT_DIR,
                           "{}-{}.jpg".format(CAL_IMAGE_PREFIX, ips.dtg()))

ips.grabImage(calFilename, isCal=True, lightType=LIGHT_TYPE)

ipd.Image(filename=calFilename)
```

Observe the calibration image above.  (You can double-click in the empty space to the left of the image to enable scroll bars.)

Does it show the entire experimental area?  Is the bottom surface (where the fish will be) in focus?  Are the edges of the experimental area aligned with the edges of the image?  Is the right side of the image upstream and the left side downstream?  If the answer to any of these is "no", then make the appropriate adjustments to the camera (and lighting) and run the cell above the image again to refresh it.

Once you are happy with the calibration image, put a fish in the flume.

Executing the next cell will perform data collection given the settings (interval, number of data, etc.) given above.

```python
###################
# DATA COLLECTION #
###################

DATA_SERIES +=1

ips.grabDataFrames(DATA_SERIES, NUMBER_OF_DATA, EXPERIMENT_DIR, DATA_IMAGE_PREFIX, INTERVAL)
ips.msg("Data series capture complete.", ips.FINAL)
```

The cell above will gather all of the data frames.  The feedback may appear at uneven intervals, but the actual data collection is performed right on time.  It will notify you when the data collection is complete.

To capture another series of data, simply run the cell again.  The series are automatically numbered and timestamped.

When you are done capturing data, proceed to the FishFace Data Analysis notebook.
