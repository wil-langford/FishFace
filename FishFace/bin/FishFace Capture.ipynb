# FishFace

This is the data capture interface for FishFace.  Use Shift-Enter to execute each cell, then look at the instructions and comments below the output.

```python
############
# SETTINGS #
############
%run ipyinit.py

# Species.  If it is set to None, then it defaults to 'HP'.
SPECIES = None

# Fish ID. If it is set to None, then it generates a new fish ID based on the date and time.
FISH_ID = None

# Treatment. If it is set to None, then it defaults to 'NT'.
TREATMENT = None

# Can be any combination of letters, numbers and spaces.  No punctuation or other special characters, please.
EXPERIMENT_NAME = ips.experimentName(SPECIES, FISH_ID, TREATMENT)

# This is the number of seconds between data image captures.  It will be rounded
# to the nearest positive integer with a minimum value of 1.
INTERVAL = 10

# How many images to capture
NUMBER_OF_DATA = 90

# Can be 'IR' or 'visible'.
LIGHT_TYPE = 'IR'

######### The following can probably be left alone. #########

# Prefixes for naming the files.
CAL_IMAGE_PREFIX = 'CAL'
DATA_IMAGE_PREFIX = 'DATA'

DATA_DIR = os.path.join('.','data')
#DATA_DIR = os.path.join('.','ephemera')
EXPERIMENT_START_TIME = ips.dtg()[:-3]
EXPERIMENT_DIR = os.path.join(DATA_DIR,
                                    "{} {}".format(EXPERIMENT_NAME,
                                                   EXPERIMENT_START_TIME))
try:
    os.mkdir(EXPERIMENT_DIR)
except OSError:
    ips.msg("Experiment directory may already exist.", ips.WARN)

DATA_SERIES = 0

ips.msg("Settings processed.", ips.FINAL)
```

```python
#############################
# CAPTURE CALIBRATION IMAGE #
#############################

calFilename = os.path.join(EXPERIMENT_DIR,
                           "{}-{}.jpg".format(CAL_IMAGE_PREFIX, ips.dtg()))

ips.grabImage(calFilename, isCal=True,  lightType=LIGHT_TYPE)
ipd.Image(filename=calFilename)
```

Observe the calibration image above.  (You can click in the empty space to the left of the image to enable scroll bars.)

Does it show the entire experimental area?  Is the bottom surface (where the fish will be) in focus?  Are the edges of the experimental area aligned with the edges of the image?  Is the right side of the image upstream and the left side downstream?  If the answer to any of these is "no", then make the appropriate adjustments to the camera (and lighting) and run the cell above the image again to refresh it.

Once you are happy with the calibration image, put a fish in the flume.

Executing the next cell will perform data collection given the settings (interval, number of data, etc.) given above.

```python
#######################
# CAPTURE DATA SERIES #
#######################

VOLTAGE = 5

ips.grabDataSeries(DATA_SERIES, NUMBER_OF_DATA, EXPERIMENT_DIR, DATA_IMAGE_PREFIX, INTERVAL, VOLTAGE, LIGHT_TYPE)
ips.msg("Data series capture complete.", ips.FINAL)

DATA_SERIES +=1
```

When enough frames have been captured, scroll bars will automatically appear in the output above.  When you see the message "data series capture complete," all images have been captured.

If you want to capture another data series (e.g. if you want to capture data for another water flow speed in the flume,) just execute the cell above again.

When all data series have been captured, you can take the fish out of the flume and proceed to the FishFace Analysis notebook.
